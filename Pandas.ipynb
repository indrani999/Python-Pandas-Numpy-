# Merging and Joining:

###### pd.concat(): Combine DataFrames along rows or columns.
###### pd.merge(): Perform SQL-like joins on DataFrames.

```python
# Create two sample DataFrames
data1 = {'A': ['A0', 'A1', 'A2'], 'B': ['B0', 'B1', 'B2']}
data2 = {'A': ['A3', 'A4', 'A5'], 'B': ['B3', 'B4', 'B5']}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Concatenate DataFrames along rows (axis=0)
result_concat = pd.concat([df1, df2], axis=0)

print("Concatenated DataFrames along rows:")
print(result_concat)
```

```python
# Create two sample DataFrames
data1 = {'key': ['A', 'B', 'C'], 'value1': [1, 2, 3]}
data2 = {'key': ['B', 'C', 'D'], 'value2': [4, 5, 6]}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Inner join on the 'key' column
result_inner = pd.merge(df1, df2, on='key', how='inner')

print("Inner join result:")
print(result_inner)

# Left join on the 'key' column
result_left = pd.merge(df1, df2, on='key', how='left')

print("\nLeft join result:")
print(result_left)

# Right join on the 'key' column
result_right = pd.merge(df1, df2, on='key', how='right')

print("\nRight join result:")
print(result_right)

# Outer join on the 'key' column
result_outer = pd.merge(df1, df2, on='key', how='outer')

print("\nOuter join result:")
print(result_outer)
```

# Pivot Tables and Crosstabs:

##### pd.pivot_table(): Create a pivot table from a DataFrame.
##### pd.crosstab(): Create a contingency table of two or more factors.

```python
pivot_table = pd.pivot_table(match, index='Date', columns='Season', values='Margin', aggfunc='mean')
print(pivot_table)
```

```python
# Create a sample DataFrame
data = {
    'Date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02'],
    'Category': ['A', 'B', 'A', 'B'],
    'Value': [10, 20, 15, 25]
}

df = pd.DataFrame(data)

# Create a pivot table with 'Date' as index and 'Category' as columns, aggregating 'Value' using the mean
pivot_table = pd.pivot_table(df, index='Date', columns='Category', values='Value', aggfunc='mean')

print("Pivot Table:")
print(pivot_table)
```

```python
#CrossTab
# Create a sample DataFrame
data = {
    'Gender': ['Male', 'Female', 'Male', 'Male', 'Female'],
    'Smoker': ['Yes', 'No', 'No', 'Yes', 'No']
}

df = pd.DataFrame(data)

# Create a contingency table for 'Gender' and 'Smoker'
crosstab = pd.crosstab(df['Gender'], df['Smoker'])

print("Contingency Table:")
print(crosstab)
```
